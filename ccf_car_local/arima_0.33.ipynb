```python
import pandas as pd
pd.set_option('display.max_columns',200)
pd.set_option('display.max_rows', 200)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_squared_error
from tqdm import tqdm
from sklearn.model_selection import KFold,StratifiedKFold
from sklearn.metrics import mean_absolute_error
import lightgbm as lgb
from pmdarima.arima import auto_arima
```

```python
path = './data/'
```

```python
train_sales_data = pd.read_csv(path + 'train_sales_data.csv')
# train_sales_data=train_sales_data.reset_index()
# train_search_data = pd.read_csv(path + 'train_search_data.csv')
# train_user_reply_data = pd.read_csv(path + 'train_user_reply_data.csv')
test_sales_data = pd.read_csv(path + 'evaluation_public.csv')
# model_body = train_sales_data[['model','bodyType']].drop_duplicates()
# test_sales_data = test_sales_data.merge(model_body,how='left',on='model')
# train_sales_data.columns = ['id', 'province', 'adcode', 'model', 'bodyType', 'regYear',
#        'regMonth', 'salesVolume']
# test_sales_data.columns = ['id', 'province', 'adcode', 'model', 'regYear',
#        'regMonth', 'salesVolume','bodyType']
```

```python
train_sales_data = train_sales_data.sort_values(['regYear','regMonth'])
```

```python
resultDF = []
for adcode in tqdm(test_sales_data.adcode.drop_duplicates()):
    for model in test_sales_data.model.drop_duplicates():
        try:
            y_val = np.log(1 + train_sales_data[(train_sales_data['adcode']==adcode)&(train_sales_data['model']==model)]['salesVolume'])
            print(len(y_val))
            arima_model = auto_arima(y_val,start_p=1, max_p=9, start_q=1, max_q=9, max_d=5,
                                     start_P=1, max_P=9, start_Q=1, max_Q=9, max_D=5,
                                     m=1, random_state=2018,
                                     trace=True,
                                     seasonal=True,
                                     error_action='ignore',
                                     suppress_warnings=True)
            preds = arima_model.predict(n_periods=4)
            preds = pd.Series(preds)
            preds = np.exp(preds) - 1
            print(preds)
            for k in range(1,5):
                tmpDF = {}
                tmpDF['adcode'] = adcode
                tmpDF['model'] = model
                tmpDF['regMonth'] = k
                tmpDF['salesVolume'] = preds[k-1]   
                print(tmpDF)
                resultDF.append(tmpDF)
        except:
            pass
        
```

```python
rest = pd.DataFrame(resultDF)
result = pd.merge(test_sales_data[['id','adcode','model','regYear','regMonth']],rest,on=['adcode','model','regMonth'],how='left')
result['forecastVolum'] = result['salesVolume'].apply(lambda x: 0 if x < 0 else int(round(x)))
result[['id','forecastVolum']].to_csv('./test.csv',index=False)
```
